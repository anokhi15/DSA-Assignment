ðŸ’¡ **Question 1**

Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

**Examples:**
Input: list1 = 5->2->3->8
list2 = 1->7->4->5
Output: New list = 5->7->4->8

Input:list1 = 2->8->9->3
list2 = 5->3->6->4
Output: New list = 5->8->9->4
</aside>

````python
# Node class for the linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to create a new linked list using two input linked lists
def create_new_linked_list(list1, list2):
    # Check if either list is empty
    if list1 is None or list2 is None:
        return None

    # Initialize the new linked list
    new_list = None
    new_list_tail = None

    # Traverse both lists simultaneously
    while list1 is not None and list2 is not None:
        # Compare the values of the current nodes
        if list1.data >= list2.data:
            greater_value = list1.data
        else:
            greater_value = list2.data

        # Create a new node with the greater value
        new_node = Node(greater_value)

        # Append the new node to the new linked list
        if new_list is None:
            new_list = new_node
            new_list_tail = new_node
        else:
            new_list_tail.next = new_node
            new_list_tail = new_node

        # Move to the next nodes in both lists
        list1 = list1.next
        list2 = list2.next

    # Return the new linked list
    return new_list

# Example 1
# Create the first linked list: 5->2->3->8
list1 = Node(5)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(8)

# Create the second linked list: 1->7->4->5
list2 = Node(1)
list2.next = Node(7)
list2.next.next = Node(4)
list2.next.next.next = Node(5)

# Create the new linked list using the two input lists
new_list = create_new_linked_list(list1, list2)

# Print the new linked list
while new_list is not None:
    print(new_list.data, end="->")
    new_list = new_list.next
# Output: 5->7->4->8

# Example 2
# Create the first linked list: 2->8->9->3
list1 = Node(2)
list1.next = Node(8)
list1.next.next = Node(9)
list1.next.next.next = Node(3)

# Create the second linked list: 5->3->6->4
list2 = Node(5)
list2.next = Node(3)
list2.next.next = Node(6)
list2.next.next.next = Node(4)

# Create the new linked list using the two input lists
new_list = create_new_linked_list(list1, list2)

# Print the new linked list
while new_list is not None:
    print(new_list.data, end="->")
    new_list = new_list.next
# Output: 5->8->9->4
````

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

**Example 1:**

```
Input:
LinkedList: 
11->11->11->21->43->43->60
Output:
11->21->43->60
```

**Example 2:**
Input:
LinkedList: 
10->12->12->25->25->25->34
Output:
10->12->25->34
</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def remove_duplicates(head):
    if head is None:
        return head
    
    current = head
    
    while current.next is not None:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next
    
    return head
# Create the linked list
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)

# Remove duplicates
head = remove_duplicates(head)

# Print the modified linked list
current = head
while current is not None:
    print(current.data, end="->")
    current = current.next
````

Given a linked list of sizeÂ **N**. The task is to reverse everyÂ **k**Â nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple ofÂ *k*Â then left-out nodes, in the end, should be considered as a group and must beÂ reversed (See Example 2 for clarification).

**Example 1:**

```
Input:
LinkedList: 1->2->2->4->5->6->7->8
K = 4
Output:4Â 2Â 2Â 1Â 8Â 7Â 6Â 5
Explanation:
The first 4 elements 1,2,2,4 are reversed first
and then the next 4 elements 5,6,7,8. Hence, the
resultant linked list is 4->2->2->1->8->7->6->5.

```

**Example 2:**
Input:
LinkedList: 1->2->3->4->5
K = 3
Output:3 2 1 5 4
Explanation:
The first 3 elements are 1,2,3 are reversed
first and then elements 4,5 are reversed.Hence,
the resultant linked list is 3->2->1->5->4.
</aside>

````python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def reverse_k_nodes(head, k):
    if head is None:
        return None
    
    prev = None
    current = head
    count = 0

    # Reverse the first 'k' nodes
    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Recursively reverse the remaining nodes
    if next_node is not None:
        head.next = reverse_k_nodes(next_node, k)

    return prev

def print_linked_list(head):
    current = head
    while current is not None:
        print(current.data, end=" ")
        current = current.next
    print()

# Example usage:
# Create the linked list: 1->2->2->4->5->6->7->8
head = Node(1)
head.next = Node(2)
head.next.next = Node(2)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

k = 4
print("Original linked list:")
print_linked_list(head)

head = reverse_k_nodes(head, k)

print("Reversed linked list:")
print_linked_list(head)
````

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseAlternateKNodes(head, k):
    current = head
    next_node = None
    prev = None
    count = 0

    # Reverse k nodes
    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Connect the reversed sublist to the next k nodes
    if head is not None:
        head.next = current

    # Skip k nodes
    count = 0
    while count < k - 1 and current is not None:
        current = current.next
        count += 1

    # Recursively reverse the next k nodes
    if current is not None:
        current.next = reverseAlternateKNodes(current.next, k)

    # Return the new head of the modified linked list
    return prev
# Create the input linked list: 1->2->3->4->5->6->7->8->9->NULL
node9 = ListNode(9)
node8 = ListNode(8, node9)
node7 = ListNode(7, node8)
node6 = ListNode(6, node7)
node5 = ListNode(5, node6)
node4 = ListNode(4, node5)
node3 = ListNode(3, node4)
node2 = ListNode(2, node3)
node1 = ListNode(1, node2)

# Call the function to reverse every alternate k nodes (k=3)
new_head = reverseAlternateKNodes(node1, 3)

# Traverse the modified linked list and print the values
current = new_head
while current is not None:
    print(current.val, end="->")
    current = current.next
````

ðŸ’¡ **Question 5**

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.

**Examples**:
Input:   1->2->3->5->2->10, key = 2
Output:  1->2->3->5->10
</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def delete_last_occurrence(head, key):
    if head is None:
        return None

    prev = None
    last = None
    curr = head

    while curr is not None:
        if curr.data == key:
            last = prev
        prev = curr
        curr = curr.next

    if last is None:
        return head

    last.next = last.next.next if last.next is not None else None
    return head

# Create the linked list: 1 -> 2 -> 3 -> 5 -> 2 -> 10
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(5)
head.next.next.next.next = Node(2)
head.next.next.next.next.next = Node(10)

# Print the original linked list
curr = head
while curr is not None:
    print(curr.data, end=" ")
    curr = curr.next
# Output: 1 2 3 5 2 10

# Delete the last occurrence of key = 2
head = delete_last_occurrence(head, 2)

print()

# Print the updated linked list
curr = head
while curr is not None:
    print(curr.data, end=" ")
    curr = curr.next
# Output: 1 2 3 5 10
````

Given two sorted linked lists consisting ofÂ **N**Â andÂ **M**Â nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

**Examples:**

Input:Â a: 5->10->15,Â b: 2->3->20

Output:Â 2->3->5->10->15->20

Input:Â a: 1->1, b: 2->4

Output:Â 1->1->2->4

</aside>

````python
# ListNode definition
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Creating List A
a = ListNode(5)
a.next = ListNode(10)
a.next.next = ListNode(15)

# Creating List B
b = ListNode(2)
b.next = ListNode(3)
b.next.next = ListNode(20)

# Function to merge lists
def mergeLists(a, b):
    # Create a dummy node as the head of the merged list
    dummy = ListNode()
    prev = dummy

    # Pointers for iterating through the linked lists
    ptr1, ptr2 = a, b

    # Merge the lists
    while ptr1 and ptr2:
        if ptr1.val <= ptr2.val:
            prev.next = ptr1
            ptr1 = ptr1.next
        else:
            prev.next = ptr2
            ptr2 = ptr2.next
        prev = prev.next

    # Attach the remaining nodes (if any) to the merged list
    prev.next = ptr1 if ptr1 else ptr2

    # Return the head of the merged list
    return dummy.next

# Merge the lists A and B
merged_head = mergeLists(a, b)

# Print the merged list
current = merged_head
while current:
    print(current.val, end="->")
    current = current.next
````

ðŸ’¡ **Question 7**

Given aÂ **Doubly Linked List**, the task is to reverse the given Doubly Linked List.

**Example:**
Original Linked list 10 8 4 2
Reversed Linked list 2 4 8 10
</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def reverse_doubly_linked_list(head):
    current = head
    previous = None

    while current is not None:
        next_node = current.next
        current.next = previous
        current.prev = next_node
        previous = current
        current = next_node

    # Update the head pointer to the last node
    head = previous

    return head

# Example usage
# Create the original linked list: 10 -> 8 -> 4 -> 2
head = Node(10)
node2 = Node(8)
node3 = Node(4)
node4 = Node(2)

head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3

# Reverse the linked list
reversed_head = reverse_doubly_linked_list(head)

# Print the reversed linked list: 2 -> 4 -> 8 -> 10
current = reversed_head
while current is not None:
    print(current.data, end=" ")
    current = current.next
````

ðŸ’¡ **Question 8**

Given a doubly linked list and a position. The task is to delete a node from given position inÂ a doubly linked list.

**Example 1:**

```
Input:
LinkedList = 1 <--> 3 <--> 4
x = 3
Output:1 3
Explanation:After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.

```

**Example 2:**
Input:
LinkedList = 1 <--> 5 <--> 2 <--> 9
x = 1
Output:5 2 9
</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


def deleteNode(head, position):
    # If the doubly linked list is empty, return the list as it is
    if head is None:
        return head

    # If the position is 1, update the head and set previous pointer of new head to None
    if position == 1:
        new_head = head.next
        if new_head:
            new_head.prev = None
        return new_head

    # Traverse the list until the desired position is reached
    current = head
    for _ in range(1, position):
        if current is None:
            return head
        current = current.next

    # Connect the previous node's next pointer to the next node
    if current.prev:
        current.prev.next = current.next

    # Connect the next node's previous pointer to the previous node
    if current.next:
        current.next.prev = current.prev

    # Free the memory occupied by the node to be deleted
    del current

    return head


# Example usage:
# Creating the doubly linked list 1 <--> 5 <--> 2 <--> 9
head = Node(1)
node2 = Node(5)
node3 = Node(2)
node4 = Node(9)

head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3

# Deleting the node at position 1
head = deleteNode(head, 1)

# Printing the modified list
current = head
while current:
    print(current.data, end=" ")
    current = current.next

# Output: 5 2 9
````
