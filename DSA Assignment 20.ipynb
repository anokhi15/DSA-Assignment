Given a binary tree, your task is to find subtree with maximum sum in tree.

   2Â  Â  Â  3

  / \Â  Â  / \

Output1 : 28

As all the tree elements are positive, the largest subtree sum is equal to sum of all tree elements.

Input2 :

 Â  Â  Â  1

     /Â  Â  \

  -2Â  Â  Â  3

  / \Â  Â  /Â  \

4 Â  5Â  -6 Â  2

Output2 : 7

Subtree with largest sum is :

 -2

 / \

4 Â  5

Also, entire tree sum is also 7.

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def findMaxSubtreeSum(node):
    if node is None:
        return 0
    
    left_sum = findMaxSubtreeSum(node.left)
    right_sum = findMaxSubtreeSum(node.right)
    
    current_sum = node.value + left_sum + right_sum
    
    global max_sum
    max_sum = max(max_sum, current_sum)
    
    return current_sum


def findMaxSubtreeSumWrapper(root):
    global max_sum
    max_sum = float('-inf')
    findMaxSubtreeSum(root)
    return max_sum


# Example binary tree
#        1
#       / \
#      2   3
#     / \ / \
#    4   5 7   6

# Create the nodes
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.right = Node(6)
root.right.left = Node(7)

# Find the maximum sum of a subtree
max_subtree_sum = findMaxSubtreeSumWrapper(root)

print("Maximum sum of a subtree:", max_subtree_sum)
```

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10


</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def constructBST(level_order):
    if not level_order:
        return None
    
    root = TreeNode(level_order[0])
    queue = [root]
    i = 1

    while i < len(level_order):
        current = queue.pop(0)

        if i < len(level_order):
            left_val = level_order[i]
            if left_val != None:
                left_child = TreeNode(left_val)
                current.left = left_child
                queue.append(left_child)
            i += 1

        if i < len(level_order):
            right_val = level_order[i]
            if right_val != None:
                right_child = TreeNode(right_val)
                current.right = right_child
                queue.append(right_child)
            i += 1

    return root

def inorderTraversal(root):
    if root:
        inorderTraversal(root.left)
        print(root.val, end=" ")
        inorderTraversal(root.right)

# Example usage:
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]
bst_root = constructBST(level_order)
inorderTraversal(bst_root)
```

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def can_represent_bst(arr):
    if not arr:
        return False

    root = Node(arr[0])
    queue = [root]
    i = 1

    while i < len(arr) and queue:
        current = queue.pop(0)

        if i < len(arr) and arr[i] < current.data:
            current.left = Node(arr[i])
            queue.append(current.left)
            i += 1

        if i < len(arr) and arr[i] > current.data:
            current.right = Node(arr[i])
            queue.append(current.right)
            i += 1

    # If there are still elements left in the array, they cannot be inserted in the BST.
    return i == len(arr)


# Test cases
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
arr2 = [11, 6, 13, 5, 12, 10]

print(can_represent_bst(arr1))  # Output: True
print(can_represent_bst(arr2))  # Output: False
```
