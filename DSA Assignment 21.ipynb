You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7

</aside>

```python
# TreeNode class definition
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def inorder_traversal(node, values):
    if node is not None:
        inorder_traversal(node.left, values)
        values.append(node.value)
        inorder_traversal(node.right, values)

def construct_bst(values):
    if not values:
        return None

    mid = len(values) // 2
    root = TreeNode(values[mid])

    root.left = construct_bst(values[:mid])
    root.right = construct_bst(values[mid+1:])

    return root

def convert_to_bst(root):
    values = []
    inorder_traversal(root, values)
    new_root = construct_bst(values)
    return new_root

# Example usage
# Create the binary tree
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Convert to binary search tree
new_root = convert_to_bst(root)

# Print the new binary search tree (in-order traversal)
def print_inorder(node):
    if node is not None:
        print_inorder(node.left)
        print(node.value)
        print_inorder(node.right)

print_inorder(new_root)
```

Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

Example:

Consider the following BST:

![1.png](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/f2455039-7e12-43fc-a7d3-b5be24931c1c/1.png)

**Input-1:**

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 6

node-2 = 14

**Output-1:**

The distance between the two keys = 4

**Input-2:**

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 3

node-2 = 4

**Output-2:**

The distance between the two keys = 2

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def build_bst(values):
    root = None
    for value in values:
        root = insert(root, value)
    return root

def insert(root, value):
    if root is None:
        return TreeNode(value)
    
    if value < root.value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    
    return root

def find_distance(root, node1, node2):
    def find_lca(node, n1, n2):
        if node is None:
            return None

        if node.value > n1 and node.value > n2:
            return find_lca(node.left, n1, n2)
        elif node.value < n1 and node.value < n2:
            return find_lca(node.right, n1, n2)
        else:
            return node

    def find_distance_to_lca(node, target, distance):
        if node is None:
            return -1
        
        if node.value == target:
            return distance
        
        left_distance = find_distance_to_lca(node.left, target, distance + 1)
        right_distance = find_distance_to_lca(node.right, target, distance + 1)
        
        if left_distance != -1:
            return left_distance
        else:
            return right_distance

    root = build_bst(values)
    lca = find_lca(root, node1, node2)
    distance_node1 = find_distance_to_lca(lca, node1, 0)
    distance_node2 = find_distance_to_lca(lca, node2, 0)

    return distance_node1 + distance_node2

# Example usage:
n = 9
values = [8, 3, 1, 6, 4, 7, 10, 14, 13]
node1 = 6
node2 = 14

print("The distance between the two keys =", find_distance(values, node1, node2))  # Output: 4

node1 = 3
node2 = 4
print("The distance between the two keys =", find_distance(values, node1, node2))  # Output: 2
```

ðŸ’¡ Question-3:

Write a program to convert a binary tree to a doubly linked list.

Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35

</aside>

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None

def binaryTreeToDoublyLinkedList(root):
    if not root:
        return None
    
    # Helper function to perform the conversion
    def convertToDoublyLinkedList(node):
        nonlocal prev, head

        if not node:
            return
        
        # Convert left subtree
        convertToDoublyLinkedList(node.left)
        
        # Convert current node
        current = DoublyLinkedListNode(node.val)
        
        if prev:
            prev.next = current
            current.prev = prev
        else:
            head = current
        
        prev = current
        
        # Convert right subtree
        convertToDoublyLinkedList(node.right)
    
    prev = None
    head = None
    
    convertToDoublyLinkedList(root)
    
    return head

# Test case
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

doublyLinkedListHead = binaryTreeToDoublyLinkedList(root)

# Print the doubly linked list
current = doublyLinkedListHead
while current:
    print(current.val, end=' ')
    current = current.next
```

Write a program to connect nodes at the same level.

Input:

        1

      /   \

    2      3

  /   \   /   \

4     5 6    7

Output:

1 â†’ -1

2 â†’ 3

3 â†’ -1

4 â†’ 5

5 â†’ 6

6 â†’ 7

7 â†’ -1

</aside>

```python
# Node class definition
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.next = None

# Function to connect nodes at the same level
def connect_nodes(root):
    if not root:
        return
    
    # Create a queue for BFS
    queue = []
    queue.append(root)
    
    while queue:
        # Get the size of the current level
        level_size = len(queue)
        
        # Connect nodes at the current level
        prev_node = None
        for _ in range(level_size):
            node = queue.pop(0)
            
            # Connect to the previous node
            if prev_node:
                prev_node.next = node
            
            # Update the previous node
            prev_node = node
            
            # Enqueue the left and right children of the current node
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
            
        # Connect the last node of the current level to None
        prev_node.next = None

# Function to print the connected nodes
def print_connected_nodes(root):
    if not root:
        return
    
    current = root
    while current:
        # Print the nodes at the current level
        temp = current
        while temp:
            print(temp.data, end=" â†’ ")
            temp = temp.next
        
        print("-1")  # Indicate the end of the current level
        current = current.left

# Example tree construction
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Connect nodes at the same level
connect_nodes(root)

# Print the connected nodes
print_connected_nodes(root)
```
